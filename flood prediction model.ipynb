```python
#remove outliers from dataset flood_train_data1
import numpy as np
from collections import Counter

def detect_outliers(flood_train_data1, features):
    outlier_indices = []
    for j in features:
        q1 = np.percentile(flood_train_data1[j], 25)
        q3 = np.percentile(flood_train_data1[j], 75)
        iqr = q3 - q1
        outlier_step = iqr * 1.5
        outlier_list_col = flood_train_data1[(flood_train_data1[j] < q1 - outlier_step) | (flood_train_data1[j] > q3 + outlier_step)].index
        outlier_indices.extend(outlier_list_col)
        
    outlier_indices = Counter(outlier_indices)
    multiple_outliers = list(i for i, v in outlier_indices.items() if v > 2)
    
    return multiple_outliers      
```

```python
#apply to existing dataset

#identify numeric columns in dataset
numeric_data = flood_train.select_dtypes(include=[np.number])
numeric_variables_list = list(numeric_data.columns)
flood_train.loc[detect_outliers(flood_train, numeric_variables_list)]
```

```python
#drop outliers
flood_train = flood_train.drop(detect_outliers(flood_train,numeric_variables_list), axis = 0).reset_index(drop = 1)
flood_train
```

```python
#separate features and target variable
from sklearn.model_selection import train_test_split
X, y = flood_train.drop('FloodProbability', axis=1), flood_train[['FloodProbability']]
```

```python
Y = y.values.ravel()
```

```python
import lightgbm as lgb
#build regression model
clf = lgb.LGBMClassifier(n_estimators=100)
```

```python
from sklearn import preprocessing
from sklearn import utils
#transform Y into an ingestedable array
print(Y)
print(type(Y))
print(np.unique(Y))
```

```python
#convert to categorical labels using bining
bins = np.linspace(min(Y), max(Y), num=4)  # Create 3 bins (4 edges)
Y_binned = np.digitize(Y, bins) - 1  # Convert to categorical labels

print(Y_binned)
```

```python
# Fit the classifier to the training data
clf.fit(X, Y_binned)
```

```python
print(X.shape)
```

```python
#predict on the training set
pred = clf.predict(X)
pred_prob = clf.predict_proba(X)

#output predictions
print("Predicted Labels:", pred)
print("Predicted Probabilities for the first sample:", pred_prob)
```

```python
print(pred_prob.shape)
```

```python
#calculate accuracy
from sklearn.metrics import accuracy_score, roc_auc_score

acc_score = accuracy_score(Y_binned, pred)
auc_score = roc_auc_score(Y_binned, pred_prob, multi_class='ovr')
print(f'The score on the training set is accuracy: {acc_score:0.4f} and AUC of {auc_score:0.4f}')
```
